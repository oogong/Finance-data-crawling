```python
import pandas as pd
import OpenDartReader
import json

# DART API를 사용하기 위한 초기화 (API 키가 필요합니다)
api_key = 'abf4e3703b22eb082cd3977afbbcc0a6746afd61'
dart = OpenDartReader(api_key)

# JSON 파일 로드
with open('name.json', 'r', encoding='utf-8') as file:
    company_list = json.load(file)

# 함수: 매출액 증가율 계산
def calculate_sales_growth_rate(fs, year):
    sales_row = fs[(fs['account_nm'] == '매출액') & (fs['fs_nm'] == '재무제표')]
    if sales_row.empty:
        return None
    thstrm_amount_str = sales_row['thstrm_amount'].values[0]
    frmtrm_amount_str = sales_row['frmtrm_amount'].values[0]
    thstrm_amount = float(thstrm_amount_str.replace(',', ''))
    frmtrm_amount = float(frmtrm_amount_str.replace(',', ''))
    sales_growth_rate = ((thstrm_amount / frmtrm_amount) - 1) * 100
    return sales_growth_rate

# 결과를 저장할 리스트
results = []

# 각 회사에 대해 매출액 증가율 계산
for company in company_list:
    code = company['Code']
    result = {'종목코드': code}
    for year in [2023, 2022, 2021]:
        try:
            fs = dart.finstate(code, str(year), reprt_code='11011')
            sgr = calculate_sales_growth_rate(fs, year)
            result[f'{year}_SGR'] = sgr
        except Exception as e:
            result[f'{year}_SGR'] = None
            print(f"Error processing {code} for year {year}: {e}")
    results.append(result)

# DataFrame으로 변환
df = pd.DataFrame(results)

# 컬럼 순서 정렬
df = df[['종목코드', '2023_SGR', '2022_SGR', '2021_SGR']]

# CSV 파일로 저장
df.to_csv('sales_growth_rate.csv', index=False)

print("CSV 파일이 저장되었습니다: sales_growth_rate.csv")
```

```python
import pandas as pd
import OpenDartReader
import json

# DART API를 사용하기 위한 초기화 (API 키가 필요합니다)
api_key = 'abf4e3703b22eb082cd3977afbbcc0a6746afd61'
dart = OpenDartReader(api_key)

# JSON 파일 로드
with open('name.json', 'r', encoding='utf-8') as file:
    company_list = json.load(file)

# 함수: 매출액 증가율 계산
def calculate_sales_growth_rate(fs):
    sales_row = fs[(fs['account_nm'] == '매출액') & (fs['fs_nm'] == '재무제표')]
    if sales_row.empty:
        return None
    thstrm_amount_str = sales_row['thstrm_amount'].values[0]
    frmtrm_amount_str = sales_row['frmtrm_amount'].values[0]
    thstrm_amount = float(thstrm_amount_str.replace(',', ''))
    frmtrm_amount = float(frmtrm_amount_str.replace(',', ''))
    sales_growth_rate = ((thstrm_amount / frmtrm_amount) - 1) * 100
    return sales_growth_rate

# 결과를 저장할 리스트
results = []

# 각 회사에 대해 2023년 매출액 증가율 계산
for company in company_list:
    code = company['Code']
    result = {'종목코드': code}
    print(result)
    try:
        fs = dart.finstate(code, '2023', reprt_code='11011')
        sgr = calculate_sales_growth_rate(fs)
        result['2023_SGR'] = sgr
    except Exception as e:
        result['2023_SGR'] = None
        print(f"Error processing {code} for year 2023: {e}")
    results.append(result)

# DataFrame으로 변환
df = pd.DataFrame(results)

# 컬럼 순서 정렬
df = df[['종목코드', '2023_SGR']]

# CSV 파일로 저장
df.to_csv('sales_growth_rate_2023.csv', index=False)

print("CSV 파일이 저장되었습니다: sales_growth_rate_2023.csv")
```
